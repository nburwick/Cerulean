```python
import pandas as pd
import json
import os
import requests
import numpy as np
from pprint import pprint
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi
```

```python
def build_API_data():
    file = open(os.path.join('..', 'static','Resources', 'Tornado_Tracks.geojson'),"r")
    data = json.load(file)
    df = pd.json_normalize(data['features'])
    return df, data['features']
```

```python
df, file_data = build_API_data()
```
